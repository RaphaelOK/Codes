```python
#A function for my blob statistics

def Crystal_Stat(file): #Individual crystal statistics
    data=pd.read_csv(file)
    df=data.groupby('ImageNumber')# This takes each image number column and creates a data frame from it
    group_MaxFD=df['AreaShape_MaxFeretDiameter']#This picks out my column of interest
    MaxFD=[] #MaxFD=Maximum Fereit Diameter
    for name,group in group_MaxFD:
        MaxFD.append(group)
        Ferit_Diameter=np.array(MaxFD)
    return Ferit_Diameter   
    
```

```python
def F(X):
    z=[]
    phia=np.arange(0,1.2,0.02)
    for i in range (len(phia)):
        F=phia[i]*log(phia[i]) +(1-phia[i])*log(1-phia[i]) + X*phia[i]*(1-phia[i])
        z.append(F)
    return z    
```

```python
X=np.array([0,1,2,2.5])
bb=[]
for x in X:
    ba=F(x)
    bb.append(ba)
 
```

```python
phib=np.arange(0,1.2,0.02)
#print phib

for i in bb:
    plt.plot(phib ,i,label='bw={0}'.format(X))
plt.legend(loc='best')
#plt.ylim(-1,0)
plt.show()
    
```

```python
def G(X,v):
    return [phia*log(phia) +(1-phia)*log(1-phia) + X*phia*(1-phia) for phia in v]
```
